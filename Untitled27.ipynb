```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country, on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3])
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=ratings,palette=['blue','red','orange','green','green'])
```

```python
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','green','green'])
```
